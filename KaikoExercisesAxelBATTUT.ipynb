```python
#will use this function later on 
def convert_timestamp_str_to_datetime(ts: str):
    """
    Takes any type of string timestamp and converts it to a pandas datetime format.
    :param ts: Timestamp in a string format understandable by pandas.to_datetime().
    :type ts: str
    :return:
    """
   

    return pd.to_datetime(ts)
```

What I have decided to do here is to gather some assets (in a list but can be a csv or json) and to determine in real time their respective slowsma and fastsma. These indicators will help me to determine if a specific asset could be a good buy.
```python
from binance.client import Client
import pandas as pd

client = Client('xxx','xxx')#create an account on binance and create an api key
```

```python
list_of_assets = ["ETHUSDT","BTCUSDT",'AVAXUSDT','FTMUSDT']
df = pd.DataFrame(list_of_assets)
df.columns = ['currency']
df
#df['currency'][0]
```

we simply have our dataframe of one column

```python
def gethourlydata(symbol):
  frame = pd.DataFrame(client.get_historical_klines(symbol, '1h', '25 hours ago UTC')) #1h for the fast sma, 25h for the slow sma

  frame =frame.iloc[:,:5] #only the five first columns we need
  frame.columns = ['Time','Open','High','Low','Close'] #rename it
  frame[['Open','High','Low','Close']] = frame[['Open','High','Low','Close']].astype(float)
  frame.Time = pd.to_datetime(frame.Time, unit='ms')

  return frame
```

```python
df = gethourlydata("BTCUSDT") #example

#here we retrieve the fast sma and the slow sma
def applytechnicals(df):  
  df['FastSMA'] = df.Close.rolling(7).mean()
  df['SlowSMA'] = df.Close.rolling(25).mean() #25 timesteps

applytechnicals(df)#get an example for btc (btw we are only interested in the last value our our slowsma and fastsma columns)
df
```

```python
should_i_buy_or_not = []
  
for coin in df["currency"]:
  df = gethourlydata(coin)
  applytechnicals(df)
  lastrow = df.iloc[-1] #only interested in the last row
  if lastrow.FastSMA < lastrow.SlowSMA: #bullish indicator
    should_i_buy_or_not.append(1)
  else:
    should_i_buy_or_not.append(1)

should_i_buy_or_not
```

```python


#creation of a dataframe by zipping two lists
df2 = pd.DataFrame(list(zip(list_of_assets, should_i_buy_or_not)),
               columns =['currency', 'buy_or_not_side'])
df2
```

That's it, second column is our indicator, and we can develop a bot to buy/sell the respective asset when the indicator switches to 1 or 0.
